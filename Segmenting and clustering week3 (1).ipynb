```python
!conda install -c anaconda beautifulsoup4 --yes
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
!conda install lxml==3.7.0 --yes
url='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
res=requests.get(url)
soup=BeautifulSoup(res.text,'html.parser')
```

```python
tables=soup.table
df_list=[]
for table in tables:
    df_list.append(pd.concat(pd.read_html(tables.prettify())))
df=pd.concat(df_list)
df
```

```python
df_a=df[~df['Borough'].str.contains('Not assigned')]
#df_at=df[df_a['Neighbourhood'].str.cat(sep=',')]
df_at=df.groupby('Postcode')['Neighbourhood'].apply(','.join).reset_index()
df_at
df_at=df_at.rename(columns={'Postcode':'Postal Code'})
df_at
```
